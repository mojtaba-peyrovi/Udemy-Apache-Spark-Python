```python
from pyspark import SparkConf, SparkContext
```

```python
conf = SparkConf().setMaster("local").setAppName("most-popular-superhero")
sc = SparkContext(conf = conf)
```

```python
def parseNames(line):
    fields = line.split('\"')
    return (int(fields[0]), fields[1].encode("utf-8"))
```

```python
def countofCoOccurance(line):
    elements = line.split()
    return (int(elements[0]), len(elements)-1)
```

```python
 names = sc.textFile("marvel-names.txt")
```

```python
names.collect()
```

```python
lines = sc.textFile("marvel-graph.txt")
```

```python
lines.collect()
```

```python
namesRdd = names.map(parseNames)
```

```python
pairings = lines.map(countofCoOccurance)
```

```python
pairings.collect()
```

```python
totalFriendsByCharacter = pairings.reduceByKey(lambda x,y: x + y) #aggregation on keys
```

```python
flipped = totalFriendsByCharacter.map( lambda x : (x[1], x[0]))
```

```python
mostPopular = flipped.max()
```

```python
print(mostPopular)
```

```python
mostPopularName = namesRdd.lookup(mostPopular[1])[0]
```

```python
print(str(mostPopularName) + " is the most popular movie with total of " + str(mostPopular[0]) + " co-appearances.")
```
